```python
from os import listdir
from os.path import isfile, join
files = [f for f in listdir("../Data/Raw") if isfile(join("../Data/Raw", f))]
print(files)
```

```python
csv_files = [f for f in files if f.split(".")[-1] == 'csv']
```

```python
csv_files
```

```python
queries = []
for f in csv_files:
    season = f.split("_")[-1].split(".")[0]
    queries.append("SELECT {} as Season, * from read_csv('../Data/Raw/{}')".format(season, f))
```

```python
conn.sql("CREATE TABLE ResultsRaw as {}".format(queries[0]))
queries.pop[0]
```

```python
queries = queries[1:]
```

```python
conn.sql("SELECT * FROM ResultsRaw").to_df()
```

```python
columns = {}
all_columns = set()
for f in csv_files:
    cols = conn.sql("select * from read_csv('../Data/Raw/{}') limit 1".format(f)).to_df().columns
    all_columns = all_columns.union(set(cols))
    columns[f] = cols
```

```python
len(all_columns)
```

Need to rectify column discrepency

```python
queries = []
for f, c in columns.items():
    missing = all_columns - set(c)
    season = f.split("_")[-1].split(".")[0]
    
    if len(missing) > 0:
        query_ext = ", Null as " + ", Null as ".join(list(missing))
    else:
        query_ext = ""
    
    queries.append("SELECT {} as Season, * {} from read_csv('../Data/Raw/{}', dateformat='%d/%m/%y')".format(season, query_ext, f))
    
```

```python
conn.sql("drop table ResultsRaw")
```

```python
full_query = "CREATE TABLE ResultsRaw as " + " UNION ALL ".join(queries)
```

```python
conn.sql(full_query)
```

```python
conn.sql("select * from ResultsRaw limit 1000").to_df().tail()
```

```python
conn.close()
```
